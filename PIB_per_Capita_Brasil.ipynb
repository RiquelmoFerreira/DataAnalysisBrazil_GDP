## Análise sobre o PIB per capita do Brasil

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Dataset

```python
Data_Set = pd.read_excel('Dados_Pib.xlsx')

Data_Set.head()
```

### Renomeando

```python
Data_Set.rename(columns={'Territorialidades':'Territory', 'Ano':'Year', 
                         'PIB per capita': 'GDP per capita(R$)'}, inplace=True)
Data_Set.head()
```

### Média

```python
Data_Set.groupby(by=['Territory', 'Year']).mean()
```

### Grids

```python
Grid_Charts = sns.FacetGrid(Data_Set, col='Territory', hue='Territory', col_wrap=4)
Grid_Charts = Grid_Charts.map(plt.plot,'Year','GDP per capita(R$)')

Grid_Charts = Grid_Charts.map(plt.fill_between, 'Year', 'GDP per capita(R$)', alpha=0.2).set_titles(
'{col_name} Territory'
)
Grid_Charts = Grid_Charts.set_titles('{col_name}')

Grid_Charts = Grid_Charts.fig.suptitle('Evolution of GDP per Capita per States', fontsize=18)

plt.subplots_adjust(top=0.92)
```
